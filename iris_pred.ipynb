```python
from sklearn.datasets import load_digits, load_iris
from sklearn.manifold import MDS, TSNE
from matplotlib import pyplot as plt
import matplotlib.cm as cm


x_data = df[feature_col_names].values
y_data= df[predicted_class_names].values
sc = preprocessing.StandardScaler()
#sc = preprocessing.MinMaxScaler()
x_data_1 = sc.fit_transform(x_data)


def plot_data(data, labels):
    plt.figure(figsize=(12,9))
    plt.scatter(data[:,0], data[:,1], c=["w" for _ in labels])
    for d, l in zip(data,labels):
        plt.text(d[0], d[1], str(l), fontdict={"size":12, "color":cm.Paired(l)})

def pca(vectors):
    print("PCA")
    pca = PCA(n_components=2)
    return pca.fit_transform(vectors)

def mds(vectors):
    mds = MDS(n_jobs=4)
    return mds.fit_transform(vectors)

def nmds(vectors):
    nmds = MDS(metric=False, n_jobs=4)
    return nmds.fit_transform(vectors)

def tsne(vectors):
    tsne = TSNE()
    return tsne.fit_transform(vectors)

def main():
    
    plot_data(pca(x_data_1), y_data)
    
    plot_data(mds(x_data_1), y_data)
    
    plot_data(nmds(x_data_1), y_data)
    
    plot_data(tsne(x_data_1), y_data)

if __name__ == "__main__":
    main()
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import Imputer
import scipy.stats
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
#df = pd.read_csv('./input/pd_vol_1_eli.csv')
pint =X.columns
X = scipy.stats.zscore(X)
print(pint)
X = pd.DataFrame(X,columns=pint)
#y = X.pop('CV_L_SwingDRate')
X
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
#from sklearn.impute import SimpleImputer 
class ReduceVIF(BaseEstimator, TransformerMixin):
    def __init__(self, thresh=5.0, impute=True, impute_strategy='median'):
        # From looking at documentation, values between 5 and 10 are "okay".
        # Above 10 is too high and so should be removed.
        self.thresh = thresh
        # The statsmodel function will fail with NaN values, as such we have to impute them.
        # By default we impute using the median value.
        # This imputation could be taken out and added as part of an sklearn Pipeline.
        if impute:
            self.imputer = Imputer(strategy=impute_strategy)

    def fit(self, X, y=None):
        print('ReduceVIF fit')
        if hasattr(self, 'imputer'):
            self.imputer.fit(X)
        return self

    def transform(self, X, y=None):
        print('ReduceVIF transform')
        columns = X.columns.tolist()
        if hasattr(self, 'imputer'):
            X = pd.DataFrame(self.imputer.transform(X), columns=columns)
        return ReduceVIF.calculate_vif(X, self.thresh)

    @staticmethod
    def calculate_vif(X, thresh=5.0):
        # Taken from https://stats.stackexchange.com/a/253620/53565 and modified
        dropped=True
        while dropped:
            variables = X.columns
            dropped = False
            vif = [variance_inflation_factor(X[variables].values, X.columns.get_loc(var)) for var in X.columns]
            
            max_vif = max(vif)
            if max_vif > thresh:
                maxloc = vif.index(max_vif)
                print(f'Dropping {X.columns[maxloc]} with vif={max_vif}')
                X = X.drop([X.columns.tolist()[maxloc]], axis=1)
                dropped=True
        return X
```

```python
transformer = ReduceVIF()
# Only use 10 columns for speed in this example
X_after = transformer.fit_transform(X[X.columns[-10:]])
X_after.head()
```
